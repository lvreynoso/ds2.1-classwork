```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("../datasets/Advertising.csv")
```

```python
data
```

```python
feature_cols = ['TV', 'radio', 'newspaper']
X_train = data[feature_cols][:150]
Y_train = data.sales[:150]
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
print(model.coef_)
print(model.intercept_)
```

```python
X_test = data[feature_cols][-50:]
Y_test = data.sales[-50:]
print(model.predict(X_test))
```

```python
print(r2_score(Y_test, model.predict(X_test))) # R^2 error; closer to 1 is better
```

```python
model.predict([[100, 100, 20]])
```

```python
print(mean_squared_error(Y_test, model.predict(X_test))) # Mean Squared Error; lower is better
```
