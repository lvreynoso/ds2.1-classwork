```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
                         7.042,10.791,5.313,7.997,5.654,9.27,3.1])


Y = np.array([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
                         2.827,3.465,1.65,2.904,2.42,2.94,1.3])

plt.scatter(X, Y)
plt.xlabel('Running Distance (Mile)')
plt.ylabel('Water Drinks (Litre)')
```

```python
test_prediction = list(map(lambda x: 0.7*x + 0.3, X))
plt.scatter(X,Y)
plt.plot(X, test_prediction)
```

```python
def find_error(prediction, observed):
    return sum([pow(observed[i] - prediction[i], 2) for i in range(len(observed))])
```

```python
error = find_error(test_prediction, Y)
```

```python
error
```

```python
minimum_error = error / len(Y)
minimum_error
```

```python
better_prediction = list(map(lambda x: 0.3*x + 0.5, X))
plt.scatter(X,Y)
plt.plot(X, better_prediction)
```

```python
better_error = find_error(better_prediction, Y)
better_error
```

```python
better_minimum = better_error / len(Y)
better_minimum
```

```python
milad_prediction = list(map(lambda x: 0.3*x + 0.8, X))
plt.scatter(X,Y)
plt.plot(X, milad_prediction)
```

```python
milad_error = find_error(milad_prediction, Y)
milad_error
```

```python
milad_minimum = milad_error / len(Y)
milad_minimum
```

```python
slide_prediction = list(map(lambda x: 0.25163494*x + 0.79880123, X))
plt.scatter(X,Y)
plt.plot(X, slide_prediction)
```

```python
slide_error = find_error(slide_prediction, Y)
slide_error
```

```python
slide_minimum = slide_error / len(Y)
slide_minimum
```
