```python
import pandas as pd
import numpy as np

data = pd.read_csv('../datasets/Churn_Modelling.csv')
data.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
from sklearn.tree import export_graphviz
import pydotplus
```

```python
data['Geography'].unique()
```

```python
# Convert string values to numerical values using LabelEncoder

data_encoded = data.apply(preprocessing.LabelEncoder().fit_transform)
print(data_encoded)
```

```python
# OneHotEncoder turns integers into a "binary" encoding
# e.g. we have Spain = 1, France = 2, Germany = 3
# we turn that into Spain = 001, France = 010, Germany = 100
# so one string value isn't regarded as "greater" than another

one_hot_encoder = preprocessing.OneHotEncoder(categories='auto')
X = one_hot_encoder.fit_transform(data_encoded).toarray()
print(X.shape)
for i in range(X.shape[1]):
    print(X[:, i])
```
